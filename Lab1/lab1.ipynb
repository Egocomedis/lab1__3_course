# Норвегія

## Основні поняття
[Picture](https://upload.wikimedia.org/wikipedia/commons/thumb/d/d9/Flag_of_Norway.svg/1920px-Flag_of_Norway.svg.png)

Норвегія, офіційна назва - *[Королівство Норвегія](https://ru.wikipedia.org/wiki/%D0%9D%D0%BE%D1%80%D0%B2%D0%B5%D0%B3%D0%B8%D1%8F)* - держава в Північній Європі, розташоване в західній частині Скандинавського півострова і на величезній кількості прилеглих дрібних островів, а також архіпелазі Шпіцберген, островах Ян-Маєн і Ведмежий в Північному Льодовитому океані.

Форма правління - конституційна монархія, форма державного устрою - унітарна. Чисельність населення Норвегії - 5 265 158 осіб (липень 2016, оцінка) це одна з найменш населених країн Європи. Щільність населення становить 16 чол. / Км². Однак розміщення населення вкрай нерівномірне. Площа Норвегії составляє 385 207 км².

### Завдання. Варіант 10

```python
a = b =0
n = int(input())
while (True):
  a += 1
  if(a**2 + b**2 == n):
    print("{0}^2+{1}^2={2}".format(a,b,n))
    break
  if(a**2 >= n):
    a = 0
    b += 1
  if(b**2 >= n):
    print("Число {0} не можна представити квадратами цiлих чисел".format(n))
    break
```
